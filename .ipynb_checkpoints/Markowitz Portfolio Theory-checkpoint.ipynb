```python
import numpy as np
import pandas as pd
import FinanceDataReader as fdr
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
corp = ['와이솔', '코웰패션', '코엔텍', '해마로푸드서비스', '다나와']
symbol = ['122990', '033290', '029960', '220630', '119860']
asset_num = len(corp)
```

```python
df = pd.DataFrame()

for sym in symbol :
    df[sym] = fdr.DataReader(sym,'2016')['Close']
    
df.columns = corp
df.head()
```

```python
plt.style.use('seaborn')
mpl.rc('font', family='nanumgothic')
mpl.rc('axes', unicode_minus=False)

df_graph = df/df.iloc[0]
df_graph.plot(figsize=(18,6))
```

```python
r = np.log(df) - np.log(df.shift(1))
r = r.drop(r.index[:1])
r.head()
```

```python
r_mean = r.mean()*252
r_mean
```

```python
r_cov = r.cov()*252
r_cov
```

```python
mpl.rc('font', family='nanumgothic')
mpl.rc('axes', unicode_minus=False)
sns.heatmap(r.corr(), cmap=mpl.cm.bone_r)
plt.show()
```

```python
r_result = []
cov_result = []
for i in range(asset_num):
    cov_result.append(r_cov.iloc[i,i])
    r_result.append(r_mean[i])

r_result = np.array(r_result)
cov_result = np.array(cov_result)
```

```python
r_result
```

```python
cov_result
```

```python
prets = []
pvols = []
for i in range (100000):
    np.random.seed(i)
    weights = np.random.random(5)
    weights /= np.sum(weights)
    prets.append(np.sum(weights * r_result))
    pvols.append(np.sqrt(np.dot(weights.T, np.dot(r_cov, weights))))
```

```python
prets = np.array(prets)
pvols = np.array(pvols)

plt.figure(figsize=(15,10))
plt.scatter(pvols, prets, c=prets/pvols, marker='o', cmap=mpl.cm.jet)
plt.grid(True)
plt.xlabel('expected std')
plt.ylabel('expected return')
plt.colorbar(label='Sharpe ratio')
plt.show()
```

```python
from scipy.optimize import minimize
```

```python
def objective(weights):
    return 0.5 * np.sqrt(np.dot(weights.T, np.dot(r_cov, weights)))
```

```python
def constraint1(weights):
    return sum(weights)-1
```

```python
w0 = [(1/asset_num) for i in range(asset_num)]
w0 = np.array(w0)
objective(w0)
```

```python
bound = [0,1]
bnds = [bound for i in range(asset_num)]
con1 = {'type':'eq','fun':constraint1}
cons = [con1]
cons
```

```python
sol = minimize(objective,w0,method='SLSQP',bounds = bnds, constraints = cons)
```

```python
print("Global Minimum Variance : " + str(np.sqrt(np.dot(sol.x.T, np.dot(r_cov, sol.x))).round(3)))
print("Expected Return : "+ str(sum(r_result*sol.x).round(3)))
```

```python
def max_sharp_ratio(y) :
    return -np.sum(y * r_result) / np.sqrt(np.dot(y.T, np.dot(r_cov, y)))
```

```python
con1 = {'type':'eq','fun':lambda y : np.sum(y) - 1}
cons = [con1]
```

```python
sol2 = minimize(max_sharp_ratio, w0, method = 'SLSQP', bounds = bnds, constraints = cons)
```

```python
r_sharp_max = np.sum(sol2.x * r_result).round(3)
std_sharp_max = np.sqrt(np.dot(sol2.x.T, np.dot(r_cov, sol2.x))).round(3)

print("Maximum Sharp Ratio Expected Return : " + str(r_sharp_max))
print("Maximum Sharp Ratio Expected Standard deviation : " + str(std_sharp_max))
```

```python
plt.figure(figsize=(15,10))
plt.scatter(pvols, prets, c=prets/pvols, marker='o', cmap=mpl.cm.jet)
plt.grid(True)
plt.title('Feasible Set Including Global Minimum Variance')
plt.xlabel('Expected std')
plt.ylabel('Expected return')
plt.colorbar(label='Sharpe ratio')
plt.scatter(np.sqrt(np.dot(sol.x.T, np.dot(r_cov, sol.x))), sum(r_result*sol.x), marker="*", s=1000)
plt.scatter(np.sqrt(np.dot(sol2.x.T, np.dot(r_cov, sol2.x))), sum(r_result*sol2.x), marker="*", s=800)
plt.show()
```

```python
def min_std(weight) :
    return np.sqrt(np.dot(weight.T, np.dot(r_cov, weight)))
```

```python
basic_var = list(0.2 for i in range(5))
bnds = list([0,1] for i in range(5))
```

```python
ef_ret = np.linspace(sum(r_result*sol.x),0.35,100)
ef_std = []

for r in ef_ret :
    con1 = {'type':'eq','fun': lambda x : np.sum(x) - 1}
    con2 = {'type':'eq','fun': lambda x : sum(r_result * x) - r}
    cons = [con1,con2]
    
    sol3 = minimize(min_std, basic_var, method = 'SLSQP', bounds = bnds, constraints = cons)
    ef_std.append(sol3['fun'])
ef_std = np.array(ef_std)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(ef_std,ef_ret, c = ef_ret / ef_std, s=5, linewidth=2, cmap=mpl.cm.jet)
```

```python
plt.figure(figsize=(15,10))
plt.scatter(ef_std,ef_ret, c = ef_ret / ef_std, marker='x', s=300, linewidth=2, cmap=mpl.cm.jet)
plt.scatter(pvols, prets, c=prets/pvols, marker='o', cmap=mpl.cm.jet)

plt.grid(True)
plt.title('Feasible Set Including Global Minimum Variance')
plt.xlabel('Expected std')
plt.ylabel('Expected return')
plt.colorbar(label='Sharpe ratio')

plt.scatter(np.sqrt(np.dot(sol.x.T, np.dot(r_cov, sol.x))), sum(r_result*sol.x), marker="*", s=1000)
plt.scatter(np.sqrt(np.dot(sol2.x.T, np.dot(r_cov, sol2.x))), sum(r_result*sol2.x), marker="*", s=800)

for n in range(len(corp)):
    plt.text(float(np.sqrt(cov_result[n])*0.98), float(r_result[n]*1.03), corp[n], fontsize=13)
    
plt.scatter(np.sqrt(cov_result), r_result, c = 'black')
plt.show()
```

```python
import scipy.interpolate as sci
```

```python
tck = sci.splrep(ef_std, ef_ret)
```

```python
def f(x):
    ''' 효율적 투자선 함수 (스플라인 근사) '''
    return sci.splev(x, tck, der=0)

def df(x):
    ''' 효율적 투자선 함수의 1차 도함수 '''
    return sci.splev(x, tck, der=1)
```

```python
plt.figure(figsize=(15,10))
plt.scatter(pvols, prets, c=prets/pvols, marker='o', cmap=mpl.cm.jet)
plt.plot(ef_std, f(ef_std), lw=7)
plt.grid(True)
plt.title('Feasible Set Including Efficient Frontier')
plt.xlabel('Expected std')
plt.ylabel('Expected return')
plt.colorbar(label='Sharpe ratio')
plt.show()
```
